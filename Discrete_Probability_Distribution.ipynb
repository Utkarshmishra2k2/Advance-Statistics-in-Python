#Bernoulli Distribution 

A Random experiment whose outcomes are of 2 tyes  i.e., Success & Failure occuring with probabilities p & q respectively is called Bernoulli Trails

A random variable X is said to have a Bernoulli distribution with parameter p if it's pmf bis given by

```
p(X = x) = (1-p)^(1-x) ; x = 0,1
     = 0 ; Otherwise
```



```
p : probability of success 
```





````python
from scipy.stats import bernoulli
````

Generating the Sample of Size 1000

````python
bernoulli.rvs(0.6, size=100)
````

Find Probability using Probability Mass Function

````python
bernoulli.pmf(k=10 , p=0.6)
````

Find Cumulative Distribution Function

````python
bernoulli.cdf(k=10 , p=0.6)
````

Moments of Distibution

````python
m , v , s , k = bernoulli.stats(0.5, moments='mvsk')
print("Mean = {} , Variance = {} , Skewness = {} , Kurtosis = {} .".format(m,v,s,k))
````

Graph Of Distribution 

````python
plt.figure(figsize=(25,25))
data_01 = bernoulli.rvs(0.6, size=1000)
plt.hist(data_01, density=True, edgecolor='black')
plt.show()
````

#Binomial Distribution 

The binomial distribution with parameters n and p is the discrete probability distribution of the number of successes in a sequence of n independent experiments, each asking a Sucess-Failure(True-False) question, and each with its own Boolean-valued outcome: success (with probability p) or failure (with probability q=1-p). 

A random variable X is said to follow Binomial Distribution if it assume only non-negatives values & it's pmf is given by;

```
P(X=x) = nCx * p^(x) * (1-p)^(n-) ;x = 0,1,2.....n
       = 0 ;Otherwise
```


    

```
n: number of trials
x: number of successes
p: probability of success on a given trial
nCk: the number of ways to obtain k successes in n trials
```



````python
from scipy.stats import binom
````

Generating the Sample of Size 1000

````python
random.binomial(n=1000, p=.5, size=100)
````

Find Probability using Probability Mass Function

````python
binom.pmf(k=10, n=12, p=0.6)
````

Find Cumulative Distribution Function

````python
binom.cdf(k=10, n=12, p=0.6)
````

Moments of Distibution

````python
m , v , s , k = binom.stats ( n = 12 , p = 0.6 , moments = "mvsk" )
print("Mean = {} , Variance = {} , Skewness = {} , Kurtosis = {} .".format(m,v,s,k))
````

Graph Of Distribution

#Pission Distribution 

The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of  space if these events occur with a known constant mean rate and independently of the time since the last event.

A discrete random variable X is said to have a Poisson distribution, with λ>0 if it has a probability mass function given by

```
P(X=k) = (λk * e– λ) / k! ;λ>0 ,x = 0,1,2......
       = 0 ; Otherwise
```



```
λ: mean number of successes that occur during a specific interval
k: number of successes
e: a constant equal to approximately 2.71828
```




````python
from scipy.stats import poisson
````

Generating the Sample of Size 100

````python
poisson.rvs(mu=10, size=100)
````

Find Probability using Probability Mass Function

````python
poisson.pmf(k=25, mu=10)
````

````python
m , v , s , k = poisson.stats ( mu=10 , moments = "mvsk" )
print("Mean = {} , Variance = {} , Skewness = {} , Kurtosis = {} .".format(m,v,s,k))
````

Graph Of Distribution 

````python
plt.figure(figsize=(25,25))
data_01 = poisson.rvs(mu=10, size=100)
plt.hist(data_01, density=True, edgecolor='black')
plt.show()
````

#Negative Binomial

The negative binomial distribution is a discrete probability distribution that models the number of failures in a sequence of independent and identically distributed Bernoulli trials before a specified (non-random) number of successes occurs

A random variable X is said to follow a negative binomial distribution with parameter r & p its pmf is given by;

```
p( X =x ) = (x+r-1)C(r-1)*p^(r)*q^(x)  ;x = 0,1,....
     = 0  ;Otherwise
```



```
x : no of failures
x + r : no of trails
p : probaility of success
```

#Geometric Distribution

Consider a sequence of trials, where each trial has only two possible outcomes (designated failure and success). The probability of success is assumed to be the same for each trial. In such a sequence of trials, the geometric distribution is useful to model the number of failures before the first success since the experiment can have an indefinite number of trials until success, unlike the binomial distribution which has a set number of trials. The distribution gives the probability that there are zero failures before the first success, one failure before the first success, two failures before the first success, and so on.

A random variable X is said to have a geometric distibution if it assumesm only non-negative valuess & it's pmf is given by

```
P(X = x ) = p*q^(x) ; x = 0,1,2.....
          = 0 ; Otherwise
```



```
x : No of failure preceding the first success
p : probability of success 
```




````python
from scipy.stats import geom
````

Generating the Sample of Size 100

````python
geom.rvs( p = 0.6 , size=100)
````

Find Probability using Probability Mass Function

````python
geom.pmf(k = 5 , p = 0.6)
````

Find Cumulative Distribution Function

````python
geom.cdf(k = 5 , p = 0.6)
````

Moments of Distibution

````python
m , v , s , k = geom.stats ( p = 0.6 , moments = "mvsk" )
print("Mean = {} , Variance = {} , Skewness = {} , Kurtosis = {} .".format(m,v,s,k))
````

Graph Of Distribution 

````python
plt.figure(figsize=(25,25))
data_01 = geom.rvs( p = 0.6 , size=100)
plt.hist(data_01, density=True, edgecolor='black')
plt.show()
````

#Hyper-Geometric Distribution

the hypergeometric distribution is a discrete probability distribution that describes the probability of k successes (random draws for which the object drawn has a specified feature) in n draws, without replacement, from a finite population of size N that contains exactly K objects with that feature, wherein each draw is either a success or a failure.

A discrete random variable X is said to follow the hypergeometricdistribution with parameter N,M & n if it assume only non-negative values & it's pmf is given by;

```
p( X = x ) = (((M)C(x))*((N-M)C(n-x))/((N)C(n)) ;x = 0,1,2....min(n,M)
           = 0;Otherwise
```



```
N: population size
M:number of success states in the population
n: number of draws (i.e. quantity drawn in each trial),
x:number of observed successes
```




````python
from scipy.stats import hypergeom
````

Generating the Sample of Size 1000

````python
hypergeom.rvs(M = 20, n = 5, N = 13 , size=100)
````

Find Probability using Probability Mass Function

````python
hypergeom.pmf(k = 4,M = 20, n = 5, N = 13)
````

Find Cumulative Distribution Function

````python
hypergeom.cdf(k = 4,M = 20, n = 5, N = 13)
````

Moments of Distibution

````python
m , v , s , k = hypergeom.stats ( M = 20, n = 5, N = 13, moments = "mvsk" )
print("Mean = {} , Variance = {} , Skewness = {} , Kurtosis = {} .".format(m,v,s,k))
````

Graph Of Distribution 

````python
plt.figure(figsize=(25,25))
data_01 = hypergeom.rvs(M = 20, n = 5, N = 13 , size=100)
plt.hist(data_01, density=True, edgecolor='black')
plt.show()
````

#Discrete Uniform Distribution

such distribution can be conveived in practice if under the given experimental condition the different value of random variable become equally likely.

A random variable is said have discrete uniform distribution over the range (low - high) if its pnf is given bt;

```
p( X = x ) = 1/( high - low ) ; low < x < high
           = 0;Otherwise
```
