## Global File


```r
library(shiny)
library(shinydashboard)
library(shinyWidgets)
library(shinyjs)
library(DT)
library(dplyr)
library(ggplot2)
#library(plotly)
file_dir <- paste0(getwd(),"/data/")
nsvdata<- read.csv(paste0(file_dir, "NSVDATA", ".csv"),header = T)
#nsvdata<-read.csv("NSVDATA.csv",header=T)
custdata<- read.csv(paste0(file_dir, "CUST_PROFILE", ".csv"),header = T)
#custdata<-read.csv("CUST_PROFILE.csv",header=T)
merged<-merge(nsvdata,custdata,by="CUSTID",all.x=T)

merged$Quarter<-cut(merged$MONTH,
                    breaks=c(0,3,6,9,12),labels=c(1,2,3,4))
```

## Server File

```r
function(input,output) {

  final_data <- reactive({
    reg_qt<-subset(merged,
                   REGION==input$Region &
                     Quarter==input$Quarter)
    nsv_bus<-aggregate(NSV~BUSINESS,data=reg_qt,FUN=sum)
    tot<-sum(nsv_bus$NSV)
    nsv_bus$share<-(nsv_bus$NSV/tot)*100
    return(nsv_bus)
  })

  output$share<-renderDataTable({

    DT::datatable(final_data(),filter = 'none', selection = "none", rownames= TRUE,
                  options = list(scrollX = FALSE,
                                 scrollY = FALSE,
                                 paging = FALSE,
                                 dom="t",
                                 columnDefs = list(
                                   list(className = 'dt-center',
                                        targets = "_all")),
                                 scrollCollapse = TRUE,
                                 tabIndex = 1,
                                 ordering = FALSE,
                                 searching = FALSE,
                                 dom = 'Bfrtip'
                  ))


  })

  output$pie<-renderPlot({
    data <- final_data()

    graph <- ggplot(data, aes(x="",y = share, fill = BUSINESS))+
      geom_bar(stat = "identity",  width=1)+
      coord_polar(theta="y", start=pi/3)+
      labs(title="Pie chart", fill="BUSINESS")
    return(graph)

  })
  output$bar = renderPlot({
    data = final_data()
    print(data)
    graph01 = ggplot(data,aes(x = BUSINESS,y = share,fill = NSV))+
      geom_bar(stat = "identity",col = "orange")
  return(graph01)
  })
}
```

```r
options(warn = -1) # suppress all warning messages generated by R


header = dashboardHeader(title = "Session1", titleWidth = 1225)

# Customizing Side-Bar Options and Drop Downs
sidebar = dashboardSidebar(disable = TRUE)

# Customizing Main-Panel and Outputs of the Dashboard
body = dashboardBody(

  # ~ Creates a logo for browser-tab.
  tags$head(
    tags$link(rel = "shortcut icon", href = "logo.ico")),

  # ~ Restricts display of error on the Dashboard. Output will be blank in case of error
  # @ Disable if you wish to see errors while developing the dashboards.
  tags$style(type="text/css", ".shiny-output-error { visibility: hidden; }",
             ".shiny-output-error:before { visibility: hidden; }"),

  tabBox(id = "tabs", height = NULL , width = 13,

         # Home Tab for Inputs
         tabPanel( id="tab1",
                   title = strong("Tab 1"), status = "primary",
                   height = NULL , width = 12,
                   fluidRow(box(
                     column(6,
                            selectInput(inputId="Region", label="Region:", choices=unique(merged$REGION), selected="North")),
                     column(6,
                            selectInput(inputId="Quarter",label="Quarter:", choices=unique(merged$Quarter), selected=1)),
                   ),
                   box(
                     column(12,
                            dataTableOutput("share")),
                   )


                   ),
                   fluidRow(box(
                     column(12,
                            plotOutput("pie"))),

                     #plotOutput("pie"),
                     box(
                     column(12,
                            plotOutput("bar")),
                   #plotOutput("bar")
                     )
                   )

         )
  )
)

# Final Step for building the UI
dashboardPage(header, sidebar, body, skin = "blue", title = "Session1")
```
